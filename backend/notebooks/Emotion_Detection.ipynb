```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import cv2
import os
```

```python
def detect_emotion(image_path):
    # Load the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Unable to load image at {image_path}. Please check the file path.")
        return None
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()

    # Analyze the image for emotion detection
    analysis = DeepFace.analyze(img, actions=['emotion'], enforce_detection=False)
    return analysis
```

```python
import os

# Example image path (update this to match your actual file location)
image_path = 'extracted_files/images/images/validation/happy/10001.jpg'  # Example image from validation set

# Check if the image file exists before proceeding
if not os.path.exists(image_path):
    print(f"Image file does not exist: {image_path}")
    # If you see this message, double-check the folder and filename spelling.
else:
    # Detect emotion
    emotion_result = detect_emotion(image_path)
    print(emotion_result)
```
